```python
# select only a part 
df = df_origin[df_origin['Name']=='GOOGL']
df.reset_index(drop= True, inplace = True)
print(df)
```

```python
# percentage change
parc =[]
for index, line in df.iterrows():
    if index == 0:
        parc.append(0)
    else:
        parc.append((line['Open']- df.loc[index-1,'Open']/line['Open'])*100)

df['parc_Open'] = parc
print(df)
```

```python
# percentage change
perc =[]
for index, line in df.iterrows():
    if index == 0:
        perc.append(0)
    else:
        perc.append((line['Volume']- df.loc[index-1,'Open']/line['Open'])*100)

df['parc_Open'] = perc
print(df)
```
